```python
with open("finaldata.txt", encoding = 'utf-8') as f:
    data = f.read()
```

```python
chunks = data.split("\n\n")
chunks = [c for c in chunks if len(c)>3]
```

```python
def parse_chunk(chunk):
    chunk = chunk.strip() #Removes extra spaces
    separated_chunk = chunk.split('\n') #Seperates/Splits all infos of an acc in a list format
    username = separated_chunk[0]
    no_of_posts = int(separated_chunk[1].split(" post")[0].replace(",",""))
    no_of_followers = float(separated_chunk[2].split(" followers")[0].replace(",","").replace("K","").replace("M",""))
    if ("K" in separated_chunk[2]):
        no_of_followers = int(no_of_followers * 1000)
    elif("M" in separated_chunk[2]):
        no_of_followers = int(no_of_followers * 1000000)
    else:
        no_of_followers = int(no_of_followers)
        
    no_of_following = float(separated_chunk[3].split(" following")[0].replace(",","").replace("K","").replace("M",""))
    if ("K" in separated_chunk[3]):
        no_of_following = int(no_of_following * 1000)
    elif("M" in separated_chunk[3]):
        no_of_following = int(no_of_following * 1000000)
    else:
        no_of_following = int(no_of_following)
        
    name = separated_chunk[4]
    if(len(separated_chunk) > 5):
        user_type = separated_chunk[5]
        bio = "\n".join(separated_chunk[6:])
    else:
        user_type = "Unknown"
        bio = " " 
    return{"Username": username, "Number Of Posts": no_of_posts, "Followers Count": no_of_followers, "Following Count": no_of_following, 
           "Name Of User": name, "Type Of User": user_type, "Bio": bio}
```

```python
all_chunks = []
for chunk in chunks:
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
print(len(all_chunks)) #Number of users in the datasets
```

```python
import json
s = json.dumps(all_chunks, indent=4)
with open("finaldata.json", "w") as f:
    f.write(s)
```

# Who has the Maximum Posts?

```python
max_value = 0
for chunk in all_chunks:
    if(max_value < (chunk["Number Of Posts"])):
        max_value = chunk["Number Of Posts"]
        chunk_with_max_posts = chunk
print(f"User with Maximum Posts: {chunk_with_max_posts['Username']} ({chunk_with_max_posts['Number Of Posts']} posts)")
```

# Who has the Maximum Followers?

```python
max_value = 0
for chunk in all_chunks:
    if(max_value < (chunk["Followers Count"])):
        max_value = chunk["Followers Count"]
        chunk_with_max_followers = chunk
print(f"User with Maximum Followers: {chunk_with_max_followers['Username']} ({chunk_with_max_followers['Followers Count']} followers)")
```

# Who follows Maximum people?

```python
max_value = 0
for chunk in all_chunks:
    if(max_value < (chunk["Following Count"])):
        max_value = chunk["Following Count"]
        chunk_with_max_following = chunk
print(f"User with Maximum Following: {chunk_with_max_following['Username']} ({chunk_with_max_following['Following Count']} following) ")
```

# Best Influence Score(Followers/Following Ratio)

```python
for chunk in all_chunks:
    chunk["Influence Score"] = round(
        chunk["Followers Count"] / chunk["Following Count"], 2
    )

chunk_with_best_influence = all_chunks[0]

for chunk in all_chunks:
    if chunk["Influence Score"] > chunk_with_best_influence["Influence Score"]:
        chunk_with_best_influence = chunk

print(f"Best influence user: {chunk_with_best_influence["Username"]} ({chunk_with_best_influence["Influence Score"]})")
```

# How many Categories are present?

```python
categories = set()
for chunk in all_chunks:
    categories.add(chunk["Type Of User"])
print(f"Number Of Categories: {len(categories)}")
```

# Average Stats of all the Parameters

```python
avg_followers = sum(chunk["Followers Count"] for chunk in all_chunks) / len(all_chunks)
avg_following = sum(chunk["Following Count"] for chunk in all_chunks) / len(all_chunks)
avg_posts = sum(chunk["Number Of Posts"] for chunk in all_chunks) / len(all_chunks)
avg_influence = sum(chunk["Influence Score"] for chunk in all_chunks) / len(all_chunks)

print(f"Average Followers: {round(avg_followers, 2)}")
print(f"Average Following: {round(avg_following, 2)}")
print(f"Average Posts: {round(avg_posts, 2)}")
print(f"Average Influence Score: {round(avg_influence, 2)}")
```
